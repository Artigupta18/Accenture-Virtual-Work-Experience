# Task 1: Data Cleaning and Modelling

```python
import pandas as pd
import numpy as np
```

# Loding and Cleaning of content dataset

```python
df_content=pd.read_csv("content.xls")
```

```python
df_content
```

```python
df_content.info()
```

```python
for column in df_content.columns:
    if df_content[column].dtype=="object":
        print(f"Column: {column}")
        print(df_content[column].value_counts(dropna=False))
        print("\n")
```

```python
df_content.duplicated().sum()
```

```python
df_content.dropna(subset=['URL'], inplace=True)
```

```python
df_content.info()
```

# Loding and Cleaning of Reaction dataset

```python
df_reaction=pd.read_csv("Reactions.xls")
```

```python
df_reaction
```

```python
df_reaction.info()
```

```python
df_reaction.dropna(subset=["Unnamed: 0","Content ID","Type","Datetime","User ID"],inplace=True)
```

```python
df_reaction.info()
```

# Loding and Cleaning of ReactionTypes dataset

```python
df_reacttype=pd.read_csv("ReactionTypes.xls")
```

```python
df_reacttype
```

```python
df_reacttype.info()
```

# Merge all dataset

```python
reaction_merged_df = pd.merge(df_reaction, df_reacttype, on='Type', how='left')
```

```python
final_merged_df = pd.merge(reaction_merged_df,df_content, on='Content ID', how='left')
```

```python
final_merged_df
```

```python
final_merged_df=final_merged_df.drop(["Unnamed: 0","User ID_x","User ID_y","Unnamed: 0_y"],axis=1)
```

```python
final_merged_df.rename(columns={'Type_x':'Reaction_type','Type_y':'Content_type'},inplace=True)
```

```python
final_merged_df.info()
```

```python
final_merged_df = final_merged_df.dropna(subset=['Category', 'Content_type'])
```

```python
final_merged_df.info()
```

# Top 5 performing categories

```python
final_merged_df['Performance Score'] = final_merged_df['Score']
category_performance = final_merged_df.groupby('Category')['Performance Score'].sum().reset_index()
top_categories = category_performance.sort_values(by='Performance Score', ascending=False).head(5)
print(top_categories)
```

# Create excel file of merged dataset and Top 5 performing categories

```python
with pd.ExcelWriter('final_output.xlsx') as writer:
    final_merged_df.to_excel(writer, sheet_name='Cleaned Dataset', index=False)
    top_categories.to_excel(writer, sheet_name='Top 5 Categories', index=False)
```
